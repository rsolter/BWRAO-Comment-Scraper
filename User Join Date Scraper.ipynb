### Reading in Users' names

```python
users = pd.read_fwf('Data/users.txt')
users = users.iloc[:, 0]
users
# Need to remove blank on far right end
```

```python

## Replaces blanks WITHIN a name with '%2520' for the URL
def url_blank_fix(string):
    return string.replace(' ','%2520')

# Removing quotes (and single trailing non-breaking space)
def remove_quotes(string): 
    return string[1:(len(string)-2)]

users = users.apply(remove_quotes)

users
```
