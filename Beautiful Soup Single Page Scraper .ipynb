```python
# Downloading a single page with requests, parsing with BeautifulSoup

# For downloading web pages
import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.blackwhitereadallover.com/2020/4/3/21206623/miralem-pjanic-marco-verratti-juventus-paris-saint-germain-2020-serie-a-summer-transfer-rumors#comments")


soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
# Getting title

print(soup.title)
#type(soup.title)

```

```python
# Getting Author Name

soup.find_all(property="author")


```

```python
# Getting the Publish Date

soup.find_all(property="article:published_time")
```
